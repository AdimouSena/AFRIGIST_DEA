# Data Structure

```python
import datacube as dcb
```

```python
dc = dcb.Datacube(app='Exercise')
```

```python
x_range= 4.5276, 4.5351
y_range = 7.5153, 7.5012
```

```python
from deafrica_tools.plotting import display_map, rgb
```

```python
display_map(x_range, y_range)
```

```python
from deafrica_tools.bandindices import calculate_indices
```

```python
from deafrica_tools.datahandling import load_ard
```

```python
ds_ls_lard = load_ard(dc, 
                      products=['s2_l2a'], 
                      min_gooddata = 0.7,
                      time=('01-2022','07-2022'),
                      x=x_range, y=y_range,
                      output_crs='EPSG:6933',
                      resolution=(-10,10),
                      measurements = ['red','blue','green','nir','swir_1']
                     )
```

```python
ds_ls_lard
```

```python
rgb(ds_ls_lard, col='time')
```

```python
ds_ls_lard = calculate_indices(ds_ls_lard, index=['NDWI','MNDWI'], satellite_mission='ls')
```

```python
ds_ls_lard
```

```python
ds_ls_lard['NDWI'].plot(col='time', col_wrap=3, cmap='Blues')
```

```python
ds_ls_lard['MNDWI'].plot(col='time', col_wrap=3)
```

```python
filtered_NDWI = ds_ls_lard.where(ds_ls_lard['NDWI'] >= 0)['NDWI']
```

```python
filtered_NDWI.plot(col='time', col_wrap=3)
```

```python
filtered_MNDWI = ds_ls_lard.where(ds_ls_lard['MNDWI'] >= 0)['MNDWI']
```

```python
filtered_MNDWI.plot(col='time', col_wrap=3)
```

```python
import geopandas as gpd
from datacube.utils import geometry
```

```python
gdf  = gpd.read_file("Data/apam.geojson")
gdf.explore()
```

```python
geom = geometry.Geometry(gdf.unary_union, gdf.crs)
```

```python
ds_l9_2 = dc.load(product='ls9_sr',
        output_crs='EPSG:6933',
        geopolygon= geom,
       resolution=(-30,30),
        time=('01-2022','07-2022'),
        measurements=['red','blue','green', 'nir', 'swir_1']
                 )
ds_l9_2
```

```python
from deafrica_tools.spatial import xr_rasterize
```

```python
mask = xr_rasterize(gdf, ds_l9_2)

    # Mask dataset to set pixels outside the polygon to `NaN`
ds_l9_2 = ds_l9_2.where(mask)
```

```python
rgb(ds_l9_2, index=1)
```
