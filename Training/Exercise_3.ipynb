# Data Structure

```python
import datacube as dcb
```

```python
dc = dcb.Datacube(app='Exercise')
```

```python
x_range= 4.5247, 4.5779
y_range = 7.5133, 7.4394
```

```python
from deafrica_tools.plotting import display_map, rgb
```

```python
display_map(x_range, y_range)
```

```python
ds_ls_lard = dc.load( product='fc_ls', 
                      time=('2000'),
                      x=x_range, y=y_range,
                      output_crs='EPSG:6933',
                      resolution=(-30,30),
                     measurements=['pv'],
                     )
```

```python
ds_ls_lard
```

```python
ds_ls_lard['pv'].isel(time=0).plot()
```

```python
ds_month = ds_ls_lard['pv'].groupby('time.month').mean()
```

```python
import matplotlib.pyplot as plt
```

```python
ds_month.sum(['x', 'y']).plot(figsize=(17,4))
plt.title('Monthly PV')
plt.ylabel('PV')
plt.xlabel('Day of month')
```

```python
ds_cond_pv = ds_ls_lard['pv'].where(ds_ls_lard['pv'] > 40)
```

```python
ds_cond_pv.isel(time=0).plot()
```

```python
ds_cond_pv.sum(['x', 'y']).plot(figsize=(17,4), marker='*', color='red')
plt.title('Monthly PV')
plt.ylabel('PV')
plt.xlabel('Day of month')
```
