# Data Structure

```python
import datacube as dcb
```

```python
dc = dcb.Datacube(app='Exercise')
```

```python
x_range= 4.5247, 4.5779
y_range = 7.5133, 7.4394
```

```python
from deafrica_tools.plotting import display_map, rgb
```

```python
display_map(x_range, y_range)
```

```python
from deafrica_tools.bandindices import calculate_indices
```

```python
from deafrica_tools.datahandling import load_ard
```

```python
ds_ls_lard = dc.load( product='rainfall_chirps_monthly', 
                      time=('2000','2021'),
                      x=x_range, y=y_range,
                      output_crs='EPSG:6933',
                      resolution=(-5000,5000)
                     )
```

```python
ds_ls_lard
```

```python
ds_ls_lard['rainfall'].isel(time=0).plot()
```

```python
import matplotlib.pyplot as plt
```

```python
ds_ls_lard['rainfall'].sum(['x', 'y']).plot(figsize=(17,4))
plt.title('Monthly rainfall')
plt.ylabel('Rainfall (mm/day)')
plt.xlabel('Day of month')
```

```python
ds_yearly = ds_ls_lard['rainfall'].groupby('time.year').sum()
```

```python
ds_yearly.mean(['x', 'y']).plot(figsize=(17,4))
plt.title('Yearly rainfall')
plt.ylabel('Rainfall (year)')
plt.xlabel('Year')
```

```python
import geopandas as gpd
from datacube.utils import geometry
```

```python
gdf  = gpd.read_file("Data/apam.geojson")
gdf.explore()
```

```python
geom = geometry.Geometry(gdf.unary_union, gdf.crs)
```

```python
ds_l9_2 = dc.load(product='ls9_sr',
        output_crs='EPSG:6933',
        geopolygon= geom,
       resolution=(-30,30),
        time=('01-2022','07-2022'),
        measurements=['red','blue','green', 'nir', 'swir_1']
                 )
ds_l9_2
```

```python
from deafrica_tools.spatial import xr_rasterize
```

```python
mask = xr_rasterize(gdf, ds_l9_2)

    # Mask dataset to set pixels outside the polygon to `NaN`
ds_l9_2 = ds_l9_2.where(mask)
```

```python
rgb(ds_l9_2, index=1)
```
